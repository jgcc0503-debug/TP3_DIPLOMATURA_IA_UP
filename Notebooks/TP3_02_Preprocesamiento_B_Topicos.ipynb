```python
# Descargar las herramientas necesarias de NLTK
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4') # Optativo pero recomendado para evitar warnings

# Inicializar el lematizador y cargar la lista de palabras vacías en inglés
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
df_train = '/content/drive/My Drive/Colab Notebooks/TP3_DIPLO/Processed/train_limpio.csv'
df_test = '/content/drive/My Drive/Colab Notebooks/TP3_DIPLO/Processed/test_limpio.csv'
```

```python
df_train = pd.read_csv(df_train)
df_test = pd.read_csv(df_test)
```

```python
def aplicar_stopwords_y_lematizacion(text):
    # 1. Asegurarnos de que sea un string
    text = str(text)

    # 2. Tokenización simple (separar la oración en una lista de palabras)
    palabras = text.split()

    # 3. Filtrar stopwords y aplicar lematización al mismo tiempo
    # Solo conservamos la palabra si no está en la lista de stopwords
    palabras_procesadas = [lemmatizer.lemmatize(word) for word in palabras if word not in stop_words]

    # 4. Volver a unir las palabras en un solo texto
    return ' '.join(palabras_procesadas)
```

```python
df_train['text_avanzado'] = df_train['text_limpio'].apply(aplicar_stopwords_y_lematizacion)
```

```python
df_test['text_avanzado'] = df_test['text_limpio'].apply(aplicar_stopwords_y_lematizacion)
```
